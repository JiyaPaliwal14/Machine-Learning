```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
ds = pd.read_csv('fruits.csv')
X = ds.iloc[:, :-1]
y = ds.iloc[:, -1]

le_color = LabelEncoder()
X['color'] = le_color.fit_transform(X['color'])

le_label = LabelEncoder()
y = le_label.fit_transform(y)
```

```python
model = DecisionTreeClassifier(criterion='gini', max_depth=3)
model.fit(X, y)
```

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=X.columns, class_names=le_label.classes_, filled=True)
plt.show()
```

```python
sample = pd.DataFrame({'color':[le_color.transform(['yellow'])[0]], 'diameter': [3]})
pred = le_label.inverse_transform(model.predict(sample))
print(pred[0])
```
